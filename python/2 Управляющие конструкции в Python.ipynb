## Условие с elif

```python
temperature_str = input()
temperature = int(temperature_str)
if temperature < -40:
    print("А может лучше дома посидим?")
else:
    if temperature < 0:
        print("Вам стоит надеть шапку")
    else:
        print("Можете отправляться без шапки")
```

## Цикл for

Давайте выведем числа от 0 до 9 включительно

```python
for i in range(1, 9, 2):
    print(i)
```

```python
for i in range(10):
    print(i)
```

```python
for temperature in range(-50, 50):
    if temperature < -40:
        print(f"Температура за окном {temperature}. А может лучше дома посидим?")
    elif temperature < 0:
        print(f"Температура за окном {temperature}. Вам стоит надеть шапку")
    else:
        print(f"Температура за окном {temperature}. Можете отправляться без шапки")
```

```python
for temperature in range(-50, 51, 10):
    if temperature < -40:
        print(f"Температура за окном {temperature}. А может лучше дома посидим?")
    elif temperature < 0:
        print(f"Температура за окном {temperature}. Вам стоит надеть шапку")
    else:
        print(f"Температура за окном {temperature}. Можете отправляться без шапки")
```

### Пропуск итерации

Напишем код

```python
i = 0
while i < 10:
    if i == 5:
        i += 1
        continue
    print(i)
    i += 1
```

```python
request = ""
while request != "стоп":
    request = input()
    if request == "пропустить":
        continue
    print("Вы ввели:", request)
```

### Обработка исключений

```python
temperature_str = input()
temperature = float(temperature_str)
if temperature < 0:
    print("Вам стоит надеть шапку")
```

```python
temperature_str = input()
try:
    temperature = float(temperature_str)
    if temperature < 0:
        print("Вам стоит надеть шапку")
except:
    print("Требуется ввести число")
```

```python
temperature_str = input("Введите температуру\n")

try:
    temperature = float(temperature_str)
    if temperature < -40:
        print(f"Температура за окном {temperature}. А может лучше дома посидим?")
    elif temperature < 0:
        print(f"Температура за окном {temperature}. Вам стоит надеть шапку")
    else:
        print(f"Температура за окном {temperature}. Можете отправляться без шапки")

except ValueError:
    print("Требуется ввести число.")
```
