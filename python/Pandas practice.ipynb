```python
def load_temperature_data(path = "lesson_data/tmp.txt"):
    """Функция для загрузки данных"""
    df = pd.read_table(path, header=None, delimiter=";")
    df = df.rename({
              0: 'station_id', 
              1: 'year',
              2: 'Jan',
              3: 'Feb',
              4: 'Mar',
              5: 'Apr',
              6: 'May',
              7: 'Jun',
              8: 'Jul',
              9: 'Aug',
              10: 'Sep',
              11: 'Oct',
              12: 'Nov',
              13: 'Dec'
        }, axis='columns')
    months = df.columns[2:]
    df[months] = df[months].apply(pd.to_numeric, errors='coerce')
    return df, months
```

## Группировка

```python
df_late
```

```python
grouped_by_station = df_late.groupby(['station_id']).mean()
grouped_by_station
```

```python
for station_id in all_stations_id:
    month_temperature = grouped_by_station.loc[station_id].values[1:]
    plt.plot(months, month_temperature)
plt.xlabel("Месяц")
plt.ylabel("Температура")
plt.title(f"Среднее значение температуры для каждой станции")
plt.show()
```

```python
for station_id in all_stations_id[::50]:
    month_temperature = grouped_by_station.loc[station_id].values[1:]
    plt.plot(months, month_temperature, label=station_id)

# plt.rcParams["figure.figsize"] = (15, 5)
plt.legend(loc="upper left")
plt.show()
```
